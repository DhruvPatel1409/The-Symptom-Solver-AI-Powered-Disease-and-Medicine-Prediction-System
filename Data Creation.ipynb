```python
import pandas as pd
pd.set_option('display.max_columns', None)
# Load the CSV file
data = pd.read_csv('medical data.csv')

# Function to find columns with value 1 in each row
def columns_with_one(row):
    # Get column names where the value is 1
    return list(row[row == 1].index)

# Apply the function to each row
data['Columns_with_1'] = data.apply(columns_with_one, axis=1)

# Show the first few rows to verify the result
data[['prognosis', 'Columns_with_1']].head(20)
```

```python
df = pd.read_csv('medical data.csv')

# Initialize an empty list to store the results
result = []

# Loop over each row in the DataFrame
for index, row in df.iterrows():
    # Find the column names where the value is 1 in the current row
    columns_with_ones = [col for col in df.columns if row[col] == 1]
    
    # Join the column names as a string and append it to the result list
    result.append(", ".join(columns_with_ones))

# Print the result
for i, res in enumerate(result):
    print(f"Row {i}: {res}")
```
